```python
def getBinary():
    binary=input("Enter binary:")
    bits=[bit for bit in binary]
    for bit in bits:
        if bit!="0" and bit!="1":
            binary=input("Invalid Bit Pattern! Please re-enter binary:")
            bits=[bit for bit in binary]
            continue
    bits=[int(bit) for bit in bits]
    return bits
```

```python
def get2dMatrix():
    rows=int(input("Enter the no of the rows: "))
    cols=int(input("Enter the no of the columns: "))
    noOfBits=rows*cols
    print(f"Enter noOfBits = {noOfBits}")
    binary=getBinary()
    while len(binary)!=noOfBits:
        print(f"Invalid input!Enter the no of bits= {noOfBits}")
        binary=getBinary()
    matrix=[]
    bit=0
    for i in range(rows):
        row=[]
        for j in range(cols):
            row.append(binary[bit])
            bit+=1
        matrix.append(row)
    return matrix
```

```python
def TwoDParityA():
    print("Sender side!----Enter the matrix so it can be send by adding parity bits")
    matrix=get2dMatrix()
    print("Orignal Input:")
    for row in matrix:
        print(row)
    for row in matrix:
        count=row.count(1)
        if count%2==1:
            row.append(1)
        else:
            row.append(0)
    columnParity=[]
    for i in range(len(matrix[0])):
        column=[row[i] for row in matrix]
        if column.count(1)%2==1:
            columnParity.append(1)
        else:
            columnParity.append(0)
    matrix.append(columnParity)
    print("Actual bits transmitted after adding parity bits: ")
    for row in matrix:
        print(row)
```

```python
def TwoDParityB():
    print("Reciever Side!----Enter the matrix having parity bits so any potential error can be detected")
    matrix=get2dMatrix()
    rows = len(matrix)
    corruptedRows=[]
    corruptedCols=[]
    for r in range(0,rows-1):
        row = matrix[r]
        if row.count(1)%2 == 1:
            print(f"Error detected in row {r}")
            corruptedRows.append(r)
    cols=len(matrix[0])
    for c in range(cols):
        column=[row[c] for row in matrix]
        if column.count(1)%2 == 1:
            print(f"Error detected in column {c}")
            corruptedCols.append(c)
    noOfRowErrors=len(corruptedRows)
    noOfColsErrors=len(corruptedCols)
    if noOfRowErrors == 1 and noOfColErrors == 1:
        errorRow=corruptedRows[0]
        errorCol=corruptedCols[0]
        if matrix[errorRow][errorCol] == 1:
            matrix[errorRow][errorCol]=0
        else:
            matrix[errorRow][errorCol] = 1
        print(f"Single error detected and corrected at  the row {errorRow} and the column {errorCol}.")
    elif noOfRowErrors==0 and noOfColErrors==0:
        print("No errors detected.")
    else:
        print("Multiple errors detected")
```

# Task-04- CRC

```python
def Xor(x,y):
    xor=[]
    for i in range(0,len(x)):
        xor.append(x[i]^y[i])
    return xor
```

```python
def binaryDivison(divisor,message):
    tempDividendLen=len(divisor)
    message.extend([0] * (len(divisor)-1))
    tempDividend=[message[i] for i in range(0,tempDividendLen)]
    xor=[]
    while tempDividendLen < len(message):
        if tempDividend[0]==1:
            xor=Xor(tempDividend,divisor)
        else:
            xor=Xor(tempDividend,[0]*len(tempDividend))
        tempDividend=xor.copy()
        tempDividend.pop(0)
        tempDividend.append(message[tempDividendLen])
        tempDividendLen+=1
    return tempDividend
```

# Task-5- CheckSum

```python
def get8bitMatrix():
    binary=getBinary()
    while len(binary)%8!=0:
        binary.insert(0,0)
    matrix=[]
    for i in range(0,len(binary),8):
        row=[]
        for j in range (i,i+8):
            row.append(binary[j])
        matrix.append(row)
    return matrix,binary
```

```python
def performBinaryAddition(matrix):
    binarySum=[0]*8
    if len(matrix)>1:
        for i in range(0,len(matrix),2):
            carry=0
            for j in range(len(matrix[0])-1,-1,-1):
                colSum=0
                if i+1<len(matrix):
                    colSum=matrix[i][j] + matrix[i+1][j] + carry
                binarySum[j]=colSum%2
                carry= colSum//2
            j = len(binarySum)-1
            while carry > 0 and j >= 0:
                colSum = binarySum[j] + carry
                binarySum[j] = colSum % 2
                carry = colSum // 2
                j = j - 1
    return binarySum
```

# Task-5- a

```python
def CheckSumA():
    matrix , binary = get8bitMatrix()
    transmittedMessage = binary.copy()
    binarySum = performBinaryAddition(matrix)
    checkSum=[1-bit for bit in binarySum]
    print("The generated Checksum is: ",checkSum)
    transmittedMessage.extend(checkSum)
    transmittedMessageStr = ''
    for bit in transmittedMessage:
        transmittedMessageStr=transmittedMessageStr+str(bit)
    print("The transmitted message will be: ",transmittedMessageStr)
```
